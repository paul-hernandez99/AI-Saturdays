# DBSCAN  (Density-based spatial clustering of applications with noise)

DBSCAN (agrupación espacial de aplicaciones con ruido basado en densidad) es un algoritmo de agrupación no supervisado de uso común propuesto en 1996. A diferencia del algoritmo K-means, DBSCAN no necesita especificar el número de agrupaciones. Puede detectar automáticamente la cantidad de clústeres en función de sus datos y parámetros de entrada. Más importante aún, DBSCAN puede encontrar agrupaciones de formas arbitrarias que K-means no puede encontrar. Por ejemplo, un grupo rodeado por un grupo diferente.

Para llegar a esto, DBSCAN agrupa los datos de acuerdo a la distancia que tengan entre ellos basándose en la distancia Euclidiana y una cantidad mínima de puntos. Además todos aquellos puntos que tengan una densidad baja se marcarán como “Outliers”. Una vez hecha esa inicialización, DBSCAN realizará los clusters a la distancia de los puntos más cercanos de acuerdo al radio _epsilon_ de esos puntos.

image.png

## Parámetros de configuración

Cuenta con dos parámetros de configuración esenciales:

- **Eps ($\epsilon$):** Radio máximo del vecino más cercano
- **minPts:** Número mínimo de puntos en un vecindario Eps de ese punto

image.png

## Otras 3 características adicionales:

- **Core point**: los puntos rojos son puntos centrales que tienen al menos MinPts = 4 en su vecindario.

- **Border point:** un punto seleccionado que se encuentra dentro de una vecindad de un punto central, pero no puede ser un punto central en sí mismo. En la figura los puntos amarillos son fronterizos.

- **Noisy point:** un punto seleccionado que no es un punto central ni un punto fronterizo. Significa que estos puntos son valores atípicos que no están asociados con ningún grupo denso. En la figura, el punto azul se identifica como punto de ruido.

image.png

image.png

## Algoritmo DBSCAN:

1. Selecciona un punto $p$ arbitrario.
2. Recupera todos los puntos de densidad alcanzables desde $p$ basado en Eps y MinPts.
3. Si $p$ es un punto central, se forma un grupo.
4. Si $p$ es un punto fronterizo, no se puede alcanzar la densidad de ningún punto desde $p$ y DBSCAN visita el siguiente punto de la base de datos.
5. Continúa el proceso hasta que se hayan procesado todos los puntos.

## Implementación en python

```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
import matplotlib
```

```python
# Creacion de funcion de datos random. IMPORTANTE: Para asegurar la reproducibilidad de las pruebas y que 
# todos podais ver la misma distribucion de puntos al generar puntos pseudoaleatoriamente, se utiliza la funcion 
# np.random.seed(<NUMERO>) con el mismo numero como parametro para que la funcion np.random se comporte 
# de la misma manera en todas las ejecuciones.

np.random.seed(42)

# Crea puntos tipo circulo
def PointsInCircum(r,n=100):
    return [(math.cos(2*math.pi/n*x)*r+np.random.normal(-30,30),math.sin(2*math.pi/n*x)*r+np.random.normal(-30,30)) for x in range(1,n+1)]
```

```python
df=pd.DataFrame(PointsInCircum(500,1000))
df=df.append(PointsInCircum(300,700))
df=df.append(PointsInCircum(100,300))

# Insertando ruido en el dataset
df=df.append([(np.random.randint(-600,600), np.random.randint(-600,600)) for i in range(300)])
```

```python
#  Visualizacion inicial

plt.figure(figsize=(10,10))
plt.scatter(df[0],df[1],s=15,color='grey')
plt.title('Dataset',fontsize=20)
plt.xlabel('Feature 1',fontsize=14)
plt.ylabel('Feature 2',fontsize=14)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
dbscan=DBSCAN()
dbscan.fit(df[[0,1]])
```

```python
# Entender la cantidad optima de puntos minimos centrales
from sklearn.neighbors import NearestNeighbors
neigh = NearestNeighbors(n_neighbors=2)
nbrs = neigh.fit(df[[0,1]])
distances, indices = nbrs.kneighbors(df[[0,1]])
```

```python
# Graficar el k-graph para ver distancias optimas. El K-distance graph nos permite entender si los epsilon estan
# dando las distancias optimas para todos los puntos.
distances = np.sort(distances, axis=0)
distances = distances[:,1]
plt.figure(figsize=(20,10))
plt.plot(distances)
plt.title('K-distance Graph',fontsize=20)
plt.xlabel('Data Points sorted by distance',fontsize=14)
plt.ylabel('Epsilon',fontsize=14)
plt.show()
```

```python
dbscan_opt = DBSCAN(eps=30, min_samples=6)
dbscan_opt.fit(df[[0,1]])
```

```python
df['DBSCAN_opt_labels']=dbscan_opt.labels_
df['DBSCAN_opt_labels'].value_counts()
```

```python
# Plotting the resulting clusters
colors = ["red", "green", 'yellow']
plt.figure(figsize=(10,10))
plt.scatter(df[0],df[1],c=df['DBSCAN_opt_labels'], cmap=matplotlib.colors.ListedColormap(colors),s=15)
plt.title('DBSCAN Clustering',fontsize=20)
plt.xlabel('Feature 1',fontsize=14)
plt.ylabel('Feature 2',fontsize=14)
plt.show()
```

# Mean Shift

Mean Shift es un algoritmo de agrupamiento por densidad. Como otro algoritmo de aprendizaje no supervisado, el procedimiento intenta agrupar datos sin haber sido entrenados primero en datos etiquetados.

Se utiliza para ello el método de **“Kernel Density Estimation"**.

Con la función de probabilidad de densidad para una porción del set de datos completo, lo que hace es seleccionar un rango de espacio y de este espacio va obteniendo el valor medio o promedio de cada uno de los puntos y va incrementando dicho valor medio, incluyendo más puntos de datos y así ir ampliando su rango de tamaño. 

Si Utilizamos un Kernel o Filtro Gausiano entonces veremos algo así como la imagen. Lo que nos indica es la densidad de los datos en diferentes puntos.

Debido a que utiliza la densidad no necesita definir los clusters requeridos, y esto hace que su coste de procesamiento sea mayor.

## Implementación en python

```python
#importar las librerias

import numpy as np
import pandas as pd
from sklearn.cluster import MeanShift
from sklearn.datasets import make_blobs  # Generador de datos con varianza
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
clusters = [[1,1,1],[5,5,5],[3,10,10]]
X, _ = make_blobs(n_samples = 150, centers = clusters, cluster_std = 0.60) 
# Retorna 2 resultados: X es la localizacion de cada punto y _ es la clasificacion de cada punto en su cluster. 
```

```python
ms = MeanShift()
ms.fit(X)
cluster_centers = ms.cluster_centers_  # se cambia de variable para poder hacer el plot. 
```

```python
fig = plt.figure(figsize=(20 , 20))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,0], X[:,1], X[:,2], marker='o')  # Dibujando los puntos
ax.scatter(cluster_centers[:,0], cluster_centers[:,1], cluster_centers[:,2], marker='x', color='red', s=300, linewidth=5) #dibujando los centros. 
plt.show()
```
